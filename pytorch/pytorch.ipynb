```python
import torch
import numpy as np
```

```python
torch.manual_seed(42)
np.random.seed(42)

def true_function(x, a):
    p_star = len(a) - 1
    return sum(a[k]/(1+2*k*torch.exp(-(x-k+4))) for k in range(p_star +1))

a = torch.tensor([3,-7,11,-7,-4,7],dtype=torch.float32)
```

```python
import torch

def dataset_gen(n, a, sigma):
    x = torch.empty(n).uniform_(-5, 5) 
    epsilon = torch.randn(n) 
    y = true_function(x, a) + sigma * epsilon
    return x, y

sigma = 0.4

n_train = 100
x_train, y_train = dataset_gen(n_train, a, sigma)

n_val =40
x_val, y_val = dataset_gen(n_val, a, sigma)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train, y_train, color="blue", label="Train data (true)", alpha=0.7)
plt.show()
```
