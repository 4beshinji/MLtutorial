# 一様乱数の生成
演習　110
一様分布 U(-1,+1)に従う乱数を10,000点生成し、その頻度分布を確率密度関数と重ねてプロットしろ。

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Seabornスタイル設定
sns.set(style="whitegrid")

# 一様分布 loc=-1, scale=2 → 区間[-1, 1)
xs = stats.uniform.rvs(loc=-1, scale=2, size=10000)

# 理論PDF用のx軸と値
x = np.linspace(-2, 2, 1000)
p = stats.uniform.pdf(x, loc=-1, scale=2)

# 平均・分散計算
v = np.var(xs)
m = np.mean(xs)

# プロット準備
plt.figure(figsize=(8, 5))
sns.histplot(xs, bins=20, stat='density', kde=False, color='skyblue', edgecolor='black', alpha=0.6, label='Histogram')

# PDFの重ね描き
plt.plot(x, p, 'r-', lw=2, label='Theoretical PDF')

# ラベルや装飾
plt.xlabel('x')
plt.ylabel('Density')
plt.title(f'Uniform Distribution: mean={m:.2f}, var={v:.3f}')
plt.legend()
plt.grid(True)
plt.tight_layout()

# 表示
plt.show()
```
