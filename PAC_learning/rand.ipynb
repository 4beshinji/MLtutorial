# 一様乱数の生成
演習　110
一様分布 U(-1,+1)に従う乱数を10,000点生成し、その頻度分布を確率密度関数と重ねてプロットしろ。

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Seabornスタイル設定
sns.set(style="whitegrid")

# 一様分布 始点loc=-1 幅scale=2 → 区間[-1, 1)
xs = stats.uniform.rvs(loc=-1, scale=2, size=10000)

# 理論PDF用のx軸と値
x = np.linspace(-1.5, 1.5, 10000)
p = stats.uniform.pdf(x, loc=-1, scale=2)

# 平均・分散計算
v = np.var(xs)
m = np.mean(xs)

# プロット準備
plt.figure(figsize=(8, 5))
sns.histplot(xs, bins=20, stat='density', kde=False, color='skyblue', edgecolor='black', alpha=0.6, label='Histogram')

# PDFの重ね描き
plt.plot(x, p, 'r-', lw=2, label='Theoretical PDF')

# ラベルや装飾
plt.xlabel('x')
plt.ylabel('Density')
plt.title(f'Uniform Distribution: mean={m:.2f}, var={v:.3f}')
plt.legend()
plt.grid(True)
plt.tight_layout()

# 表示
plt.show()
```

# 離散乱数の生成
離散乱数x in {1,2,3} が次の確率で生起するとする：
　P(X=x)={(x+2)/12 if x in {1,2,3}
         {0        if x not in {1,2,3}

上記の離散分布に従って100点生成し、実現値1,2,3の割合を、真の割合と合わせてプロットしろ

```python
import numpy as np
import matplotlib.pyplot as plt

# 定義された確率分布
x_values = np.array([1, 2, 3])
true_probs = np.array([(x + 2) / 12 for x in x_values])  # [(1+2)/12, (2+2)/12, (3+2)/12] = [0.25, 0.333, 0.417]

# 乱数生成（100点）
samples = np.random.choice(x_values, size=100, p=true_probs)

# 実現値の割合を計算
sample_counts = np.array([np.sum(samples == x) for x in x_values])
sample_probs = sample_counts / 100

# プロット
bar_width = 0.35
indices = np.arange(len(x_values))

plt.figure(figsize=(8, 5))
plt.bar(indices, true_probs, width=bar_width, label='True Probabilities', color='skyblue')
plt.bar(indices + bar_width, sample_probs, width=bar_width, label='Sampled Probabilities', color='orange')
plt.xticks(indices + bar_width / 2, x_values)
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('True vs Sampled Probabilities (100 samples)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
