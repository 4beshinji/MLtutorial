# 一様乱数の生成
演習　110
一様分布 U(-1,+1)に従う乱数を10,000点生成し、その頻度分布を確率密度関数と重ねてプロットしろ。

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 一様分布 loc=1, scale=1 → 区間[1,2)
xs = stats.uniform.rvs(loc=-1, scale=2, size=10000)

# 理論PDF用のx軸と値
x = np.linspace(-2, 2, 1000)
p = stats.uniform.pdf(x, loc=-1, scale=2)

# 平均・分散計算
v = np.var(xs)
m = np.mean(xs)

# プロット
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(xs, bins=10, alpha=0.5, density=True)  # ← density=True に修正
ax.plot(x, p, 'r-', lw=2)

# ラベルなど
ax.set_xlabel('x')
ax.set_ylabel('p(x)')
ax.set_title('hist mean=%.2f, var=%.3f' % (m, v))
ax.grid(True)

plt.show()
```
